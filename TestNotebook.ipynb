```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV-Datei laden
df = pd.read_csv('data.csv', parse_dates=['datum'])

# Convert datetime to date
#df['datum'] = df['datum'].dt.date

# Gruppieren nach Datum und die Anzahl der Zugriffe summiert
daily_counts = df.groupby('datum')['anzahl'].sum()

# Convert index to formatted string
daily_counts.index = daily_counts.index.strftime('%d-%m-%Y')

# Plot erstellen
fig, ax = plt.subplots(figsize=(20, 6))
ax.plot(daily_counts.index, daily_counts.values, 'b.-')
ax.set_title('Tgliche Zugriffe')
ax.set_xlabel('Datum')
ax.set_ylabel('Anzahl Zugriffe')

# Diagramm anzeigen
plt.show()
```
