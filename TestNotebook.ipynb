```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV-Datei laden
df = pd.read_csv('data.csv', parse_dates=['datum'])

# Gruppieren nach Datum und die Anzahl der Zugriffe summiert
daily_counts = df.groupby('datum')['anzahl'].sum()

# Plot erstellen
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(daily_counts.index, daily_counts.values, 'b.-')
ax.set_title('Tägliche Zugriffe')
ax.set_xlabel('Datum')
ax.set_ylabel('Anzahl Zugriffe')

# Diagramm anzeigen
plt.show()
```
